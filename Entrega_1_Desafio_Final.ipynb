```python
import pandas as pd
import requests 
import schedule
import time
```

# Extração tabela bancos

```python
url = "https://brasilapi.com.br/api/banks/v1"
response = requests.get(url)
```

```python
if response.status_code == 200:
    data_json = response.json()
    from plyer import notification 
    notification.notify(
     title='SUCESSO',
     message='Base carregada com sucesso',
     app_name='API',
     timeout=10 )
else:
    from plyer import notification 
    notification.notify(
     title='ERRO',
     message='Erro ao carregar API',
     app_name='API',
     timeout=10 )
```

```python
bancos  = pd.DataFrame(data_json)
bancos.head(10)
```
