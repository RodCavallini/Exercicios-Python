```python
import pandas as pd
import numpy as np
```

```python
fastfood = pd.read_excel("fastfood.xlsx")
fastfood.head(3)
```

```python
print("count:  ",fastfood['protein'].count())
print("maximo: ",fastfood['protein'].max())
print("minimo: ",fastfood['protein'].min())
print("mean:   ",fastfood['protein'].mean())
print("std:    ",fastfood['protein'].std())
print("var:    ",fastfood['protein'].var())
```

```python
print("median:   ",fastfood['protein'].median())
print("quantile: ",fastfood['protein'].quantile(0))#min
print("quantile: ",fastfood['protein'].quantile(0.25))
print("quantile: ",fastfood['protein'].quantile()) #median
print("quantile: ",fastfood['protein'].quantile(0.75))
print("quantile: ",fastfood['protein'].quantile(1))#max
```

```python
fastfood.median(numeric_only = True)
```

```python
fastfood.describe().round(2)
```

Group By

```python
fastfood.groupby('restaurant')['calories'].mean()
```

```python
fastfood.groupby('restaurant')[['calories']].mean()
```

```python
fastfood.groupby('restaurant').median(numeric_only=True)
```

```python
numeric_columns = fastfood.select_dtypes(include=np.number).columns
fastfood.groupby('restaurant')[numeric_columns].agg(['sum','max'])
```

```python
fastfood.groupby('restaurant').agg({'calories': 'mean', 'cal_fat': 'max'})
```

```python
fastfood.groupby('restaurant').agg({'calories': ['mean','max']})
```

Concat

```python
df1 = pd.DataFrame({'produto':['relogio','anel','brinco','camisa'],
                    'loja':['Zimer','Piter','Piter','Loyyn'],
                    'valor':[100, 70, 20, 120]})
df1
```

```python
df2 = pd.DataFrame({'produto':['gravata','colar','bolsa'],
                    'loja':['Loyyn','Piter','Piter'],
                    'valor':[60, 120, 300]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
df3 = pd.DataFrame({'cidade':['Loyyn','Piter','Zimer'],
                    'loja':['New Romy','Piter','Golgo']})
df3
```

```python
pd.concat([df1,df3], ignore_index=True)
```

```python
pd.concat([df1,df3], axis=1)
```

MERGE(JOIN)


```python
df1.merge(df3,on='loja',how='left')
```

```python
df1.merge(df3,on='loja',how='inner')
```

```python
df1.merge(df3,on='loja',how='left')
```

```python
df1.merge(df3,on='loja',how='right')
```

```python
df1.merge(df3,on='loja',how='outer')
```
