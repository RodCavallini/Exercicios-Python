```python
import pandas as pd
```

```python
df = pd.read_csv('realtor-data.csv')
df.head()
```

```python
df.shape
```

```python
forsale = df.loc[df['status'] == 'for_sale']
forsale
```

```python
df.value_counts('status')
```

```python
massa = forsale.loc[forsale['state'] == 'Massachusetts']
massa = massa.sort_values('house_size')
massa = massa.drop_duplicates(subset=['status','city','state','price'],keep='last')
massa = massa.sample(n = 250 , random_state=1)
massa
```

```python
puerto = forsale.loc[forsale['state'] == 'Puerto Rico']
puerto = puerto.sort_values('house_size')
puerto = puerto.drop_duplicates(subset=['status','city','state','price'],keep='last')
puerto = puerto.sample(n = 250 , random_state=1)
puerto
```

```python
conne = forsale.loc[forsale['state'] == 'Connecticut']
conne
```

```python
rhode = forsale.loc[forsale['state'] == 'Rhode Island']
rhode
```

```python
york = forsale.loc[forsale['state'] == 'New York']
york
```

```python
juncao = pd.concat([massa,puerto,conne,rhode,york])
juncao.sort_values('house_size')
juncao.drop_duplicates(subset=['status','city','state','price'],keep='last')
```
