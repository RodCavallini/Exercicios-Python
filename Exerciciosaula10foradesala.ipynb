1. Usando o arquivo "best-selling game consoles.xlsx" conte quantos consoles tem na base por companhias.

```python
import pandas as pd
game = pd.read_excel('best-selling game consoles.xlsx',sheet_name='consoles')
game.head(10)
```

```python
contagem = game.groupby('Company')[['Console Name']].count()
contagem
```

2 - Ordene o resultado anterior pela quantidade de consoles por companhia de forma decrescente.

```python
ordenado = contagem.sort_values('Console Name',ascending=False)
ordenado
```

3 - Sobre o dataframe anterior, adicione a coluna "Market Value" presente no tabela do arquivo "company.txt".

```python
market_value = pd.read_csv("company.txt")
market_value
```

```python
ordenado1 = ordenado.merge(market_value,on='Company',how='left')
ordenado1
```

```python
soma = ordenado1.sum()
soma
print("Valor Médio: ", soma['Market Value']/soma['Console Name'])
```
