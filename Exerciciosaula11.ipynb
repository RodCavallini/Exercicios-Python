```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fastfood = pd.read_excel("fastfood.xlsx")
fastfood.head(20)
```

```python
fastfood_agg = fastfood.groupby('restaurant').mean(numeric_only=True).reset_index()

plt.bar(fastfood_agg['restaurant'], fastfood_agg['calories'], color='blue')
plt.xticks(rotation=45)

plt.title("Media de Calorias por Restaurante")
plt.xlabel("Restaurante")
plt.ylabel("Media de Calorias")

plt.show()
```

```python
plt.hist(fastfood['calories'], bins=8)

plt.title("Histograma Calorias")


plt.show()
```

```python
# Plotar histograma e distribuição
sns.displot(fastfood['calories'], kde=True )
plt.show()
```

```python
sns.scatterplot(x='calories', y='sugar', size = 'protein',data=fastfood, alpha=0.7)
plt.show()
```
