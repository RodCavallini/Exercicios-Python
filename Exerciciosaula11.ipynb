```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fastfood = pd.read_excel("fastfood.xlsx")
fastfood.head(20)
```

```python
data = {'item' : [fastfood('item')],
        'calories': [fastfood('calories')]}
df = pd.DataFrame(data)

# Plotar scatter plot
sns.scatterplot(x='item', y='calories', data=df)
plt.show()
```
